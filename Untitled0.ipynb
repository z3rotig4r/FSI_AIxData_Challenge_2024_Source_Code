```python
import platform
import sys
import subprocess
import pkg_resources

# OS 정보
os_info = f"Operating System: {platform.system()}\n" \
          f"OS Version: {platform.version()}\n" \
          f"OS Release: {platform.release()}\n" \
          f"Machine: {platform.machine()}\n"

# Python 버전
python_info = f"\nPython Version: {sys.version}\n"

# 추가 시스템 정보
system_info = f"\nProcessor: {platform.processor()}\n" \
              f"Architecture: {platform.architecture()[0]}\n"

# GPU 정보 (if available)
gpu_info = subprocess.run(["nvidia-smi"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
gpu_info_str = "\nGPU Information:\n" + gpu_info.stdout if gpu_info.returncode == 0 else "\nNo GPU detected or NVIDIA-SMI not installed.\n"

# 설치된 패키지 목록 가져오기 (pip 사용)
installed_packages = subprocess.check_output(['pip', 'freeze']).decode('utf-8')

# 모든 정보를 파일로 저장
with open("environment_info.txt", "w") as f:
    f.write(os_info)
    f.write(python_info)
    f.write(system_info)
    f.write(gpu_info_str)
    f.write("\nInstalled Libraries and Versions:\n")
    f.write(installed_packages)

print("Environment information and installed packages have been saved to 'environment_info.txt'")
```

```python
import pkg_resources

# Installed libraries and their versions
installed_packages = pkg_resources.working_set
installed_packages_list = sorted([f"{i.key}=={i.version}" for i in installed_packages])

print("Installed Libraries and Versions:")
for package in installed_packages_list:
    print(package)

# Save to file
with open("installed_packages.txt", "w") as f:
    for package in installed_packages_list:
        f.write(f"{package}\n")

print("\nAll installed packages and their versions have been saved to 'installed_packages.txt'")
```

```python
import platform
import sys
import os
import subprocess
import pkg_resources

# OS 정보
print("Operating System:", platform.system())
print("OS Version:", platform.version())
print("OS Release:", platform.release())
print("Machine:", platform.machine())

# Python 버전
print("\nPython Version:", sys.version)

# Installed libraries and their versions
print("\nInstalled Libraries and Versions:")
for dist in sorted(pkg_resources.working_set, key=lambda x: x.project_name.lower()):
    print(f"{dist.project_name}=={dist.version}")

# Additional system information
print("\nProcessor:", platform.processor())
print("Architecture:", platform.architecture()[0])

# GPU 정보 (if available)
gpu_info = subprocess.run(["nvidia-smi"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
if gpu_info.returncode == 0:
    print("\nGPU Information:")
    print(gpu_info.stdout)
else:
    print("\nNo GPU detected or NVIDIA-SMI not installed.")
```

```python
import platform
import sys
import os
import subprocess
import pkg_resources

# Output to a file
with open('environment_info.txt', 'w') as f:
    f.write(f"Operating System: {platform.system()}\n")
    f.write(f"OS Version: {platform.version()}\n")
    f.write(f"OS Release: {platform.release()}\n")
    f.write(f"Machine: {platform.machine()}\n")

    f.write(f"\nPython Version: {sys.version}\n")

    f.write("\nInstalled Libraries and Versions:\n")
    for dist in sorted(pkg_resources.working_set, key=lambda x: x.project_name.lower()):
        f.write(f"{dist.project_name}=={dist.version}\n")

    f.write(f"\nProcessor: {platform.processor()}\n")
    f.write(f"Architecture: {platform.architecture()[0]}\n")

    # GPU 정보 (if available)
    gpu_info = subprocess.run(["nvidia-smi"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    if gpu_info.returncode == 0:
        f.write("\nGPU Information:\n")
        f.write(gpu_info.stdout)
    else:
        f.write("\nNo GPU detected or NVIDIA-SMI not installed.\n")

print("Environment information has been saved to environment_info.txt")
```
